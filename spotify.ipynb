DROPPING ENTRIES WITH MISSING VALUES/NULL VALUES, SINCE FILLING THEM BY ANY MEANS WOULD BE INAPPROPRIATE

```python
df_1 = df.copy()
df_1.dropna(inplace=True)
```

FINDING DUPLICATE ENTRIES

```python
duplicates = df_1[df_1.duplicated(keep=False)]
grouped_duplicates = duplicates.groupby(list(df_1.columns)).size().reset_index(name='Count')
print(grouped_duplicates)
```

```python
initial_row_count = len(df)
df_no_duplicates = df_1.drop_duplicates()
final_row_count = len(df_no_duplicates)
removed_duplicates_count = initial_row_count - final_row_count
print("Initial DataFrame rows:", initial_row_count)
print("Rows after eliminating duplicates:", final_row_count)
print("Duplicate rows removed:", removed_duplicates_count)

df_1 = df_no_duplicates.copy()
```

FUNCTION TO IDENTIFY OUTLIERS

```python
def identify_outliers(data):
    numeric_cols = data.select_dtypes(include=[np.number])

    # Calculate quartiles
    first_quartile = numeric_cols.quantile(0.25)
    third_quartile = numeric_cols.quantile(0.75)

    # Calculate Interquartile Range (IQR)
    iqr = third_quartile - first_quartile

    # Define lower and upper bounds for outliers
    lower_bound = first_quartile - 1.5 * iqr
    upper_bound = third_quartile + 1.5 * iqr

    # Identify outliers
    are_outliers = ((numeric_cols < lower_bound) | (numeric_cols > upper_bound)).any(axis=1)

    return data[are_outliers]

# Detect and display outlier records
outlier_records = identify_outliers(df_1)
print("Outlier records:")
print(outlier_records)
```

DISTRIBUTION PLOT FOR THE PROCESSED DATA

```python
num_columns = df_1.select_dtypes(include=['int', 'float'])
num_columns.hist(figsize=(12, 10))
plt.show()
```

CORRELATIONAL ANALYSIS OF THE PROCESSED DATA

```python
correlation_plot = num_columns.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_plot,annot=True,cmap='magma')
plt.title("Correlational heatmap")
plt.show()
```

FINDING THE TOP 20 GENRES WITH MOST SONGS

```python
songs_per_genre = df_1['genre'].value_counts()
songs_per_genre.head(20)
```

FINDING THE TOP 50 ARTISTS WITH MOST SONGS

```python
songs_per_artist = df_1['artistName'].value_counts()
plt.figure(figsize=(12, 6))
songs_per_artist[:50].plot(kind='bar',color = 'purple') 
plt.title('Top 20 Artists with most songs')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

FINDING THE TOP 10 ARTISTS WITH MOST AVERAGE LOUDNESS

```python
artist_danceability = df_1.groupby('artistName')['loudness'].mean()

top_10_artists_danceability = artist_danceability.nlargest(10)

print(top_10_artists_danceability)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_10_artists_danceability.values, y=top_10_artists_danceability.index, palette="viridis")
plt.xlabel('Average loudness (db)')
plt.ylabel('Artist Name')
plt.title('Top 10 loudest artists')
plt.tight_layout()

# Show the plot
plt.show()
```

ARTISTS WITH MOST SONGS IN A GENRE

```python
artist_genre_counts = df_1.groupby(['genre', 'artistName']).size().reset_index(name='ArtistCount')

popular_artists_by_genre = artist_genre_counts.groupby('genre').apply(lambda x: x.loc[x['ArtistCount'].idxmax()])

popular_artists_by_genre = popular_artists_by_genre.reset_index(drop=True)

print(popular_artists_by_genre[['genre', 'artistName', 'ArtistCount']])
```

ACOUSTICNESS VS ENERGY IN TOP 10 GENRES

```python
top_10_genres = df_1['genre'].value_counts().head(10).index

df_top_10_genres = df_1[df_1['genre'].isin(top_10_genres)]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_top_10_genres, x='acousticness', y='energy', hue='genre', palette='viridis')
plt.title('Acousticness vs. Energy in Top 10 Genres')
plt.xlabel('Acousticness')
plt.ylabel('Energy')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

SPEECHINESS ACROSS TOP 10 ARTISTS

```python
top_10_artists = df_1['artistName'].value_counts().head(10).index

df_top_10_artists = df_1[df_1['artistName'].isin(top_10_artists)]

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_top_10_artists, x='artistName', y='speechiness')
plt.title('Speechiness across Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Speechiness')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

TEMPO DISTRIBUTION ACROSS TOP 10 GENRES

```python
top_10_genres = df_1['genre'].value_counts().head(10).index
df_top_10_genres = df_1[df_1['genre'].isin(top_10_genres)]

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_top_10_genres, x='genre', y='tempo')
plt.title('Tempo Distribution across Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Tempo (BPM)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

KEY DISTRIBUTION ACROSS TOP 10 GENRES

```python
top_10_genres = df_1['genre'].value_counts().head(10).index

df_top_10_genres = df_1[df_1['genre'].isin(top_10_genres)]

plt.figure(figsize=(10, 6))
sns.countplot(data=df_top_10_genres, x='key', hue='genre')
plt.title('Key Distribution across Top 10 Genres')
plt.xlabel('Key')
plt.ylabel('Count')
plt.legend(title='Genre', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
